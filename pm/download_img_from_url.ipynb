```python
from PIL import Image
import requests
from io import BytesIO
import json
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from skimage.filters import threshold_otsu
from skimage.io import imread, imsave
from skimage.color import rgb2gray
from skimage.measure import label, regionprops
import random
import math
import imageio
import imgaug as ia
from imgaug import augmenters as iaa
from imgaug.augmentables.bbs import BoundingBox, BoundingBoxesOnImage
import glob
import random
import shutil
```

```python
# get image from Pm_ext.json (training set)
count = 1
species = {
    "negative": len(glob.glob("../datasets/train/negative/negative-*.jpg"))+1,
    "pf": 1,
    "pm": 1,
    "po": 1,
    "pv": 1
}

specie = "pm"
phase = "train"
degs = [45,90,135,180,225,270,315,360]

data = (requests.get('https://raw.githubusercontent.com/RTANC/malaria_dataset/master/Malariae/Pm_ext.json')).json()
images = data["_via_img_metadata"]
for key in images:
    url = images[key]["filename"]
    leng = len(images[key]["regions"])
    if leng > 0:
        print("{} {}".format(count,url))
        response = requests.get(url)
        img = Image.open(BytesIO(response.content))
        w, h = img.size
        for box in images[key]["regions"]:
            lbl = box["region_attributes"]["type"]
            b = box["shape_attributes"]
            img_crop = img.crop((b["x"],b["y"],(b["x"] + b["width"]),(b["y"] + b["height"])))
            if lbl == "1":
                name = specie + "-" + phase + "-{:010d}".format(species[specie])
                img_crop.save(os.path.join('../datasets',phase,specie,name+".jpg"))
                print(name)
                species[specie] += 1
                img_crop = np.array(img_crop)
                for deg in degs:
                    image_aug = iaa.Affine(rotate=deg,mode="reflect")(image=img_crop)
                    image_aug = Image.fromarray(image_aug)
                    name = specie + "-" + phase + "-{:010d}".format(species[specie])
                    image_aug.save(os.path.join('../datasets',phase,specie,name+".jpg"))
                    print(name)
                    species[specie] += 1
            elif lbl == "0":
                name = "negative-" + phase + "-{:010d}".format(species["negative"])
                img_crop.save(os.path.join('../datasets',phase,"negative",name+".jpg"))
                print(name)
                species["negative"] += 1
        count += 1
```
